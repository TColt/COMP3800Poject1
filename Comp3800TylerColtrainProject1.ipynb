```python
df_totals = df.groupby(['Region']).sum()
df_totals.drop(labels='Total', axis=1, inplace=True)
df_totals
```

```python
df_region_change= df_totals.transpose()
```

```python
df_region_change.columns
```

```python
fig, ax = plt.subplots()
ax.set_ylabel('Metric tons per capita')
ax.set_xlabel('Year')
df_region_change.plot(kind='line', ax=ax, figsize=(15,10), title='Change in CO2 emissions per region 1990 - 2019')
```
